```python
import requests
import pandas as pd
from datetime import datetime

# API key
API_KEY = "0ad293d76eeb6d22c16e891eb48b3708"

# Ciudad
city = "Monterrey"
url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric&lang=en"

# Solicitud
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    
    # Extraer datos clave
    weather = data["weather"][0]["description"]
    temp = data["main"]["temp"]
    feels_like = data["main"]["feels_like"]
    humidity = data["main"]["humidity"]
    wind_speed = data["wind"]["speed"]
    timestamp = datetime.utcnow()  # hora en UTC
    
    # Crear DataFrame
    df = pd.DataFrame([{
        "city": city,
        "timestamp": timestamp,
        "weather": weather,
        "temp_C": temp,
        "feels_like_C": feels_like,
        "humidity_%": humidity,
        "wind_speed_m_s": wind_speed
    }])
    
    # Guardar en CSV
    df.to_csv("weather_data.csv", index=False, mode="a", header=not pd.io.common.file_exists("weather_data.csv"))
    
    # Guardar en Parquet
    df.to_parquet("weather_data.parquet", index=False)
    
    print("Datos guardados en CSV y Parquet âœ…")
    print(df)
else:
    print("Error fetching data:", response.status_code, response.text)
```
